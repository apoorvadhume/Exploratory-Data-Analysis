## Inspecting null values

```python
#finding the percentage of null values in each column in the updated dataframe. All should be less than 50%
for i in range (0,121):
    print(i, df.iloc[0:0,i], '\n Percentage of null values: ', round((1-df.iloc[:,i].isnull().value_counts(normalize=True)[False])*100,2),'% \n')
```

```python
#finding which columns have more than 50% null values and saving them in a list

j=0
cols_to_drop=[]
for i in range (0,121):
    if round((1-df.iloc[:,i].isnull().value_counts(normalize=True)[False])*100,2) > 50:
        print(i, df.columns[i], '\n Percentage of null values: ', round((1-df.iloc[:,i].isnull().value_counts(normalize=True)[False])*100,2),'% \n')
        cols_to_drop.append(df.columns[i])
        j = j+1
print('Total columns to drop:',j)
print(cols_to_drop)
```

```python
#dropping those columns having more than 50% null values

df.drop(cols_to_drop, axis=1, inplace=True)
```

Instead of capping  or dropping outliers, there is one more way to deal with these outliers. Throughout the analysis we can avoid using mean and can look at median/ quantiles/ percentiles, as they are not affected by outliers.

```python
# binning income, credit ammount and annuity into 5 bins, so as to use these columns by exclusing outliers.
df['AMT_INCOME_TOTAL_BINS'] = pd.qcut(df['AMT_INCOME_TOTAL'], 5, labels=['Very Low(<99,000)','Low(99,0000-1,20,000)','Mid(1,20,000-1,70,000)','High(1,70,000-2,50,000)','Very High(>2,25,000)'])
df['AMT_CREDIT_BINS'] = pd.qcut(df['AMT_CREDIT'], 5, labels=['Very Low(<80000)', 'Low (80000-1,50,000)', 'Mid (150000-350000)', 'High (350000-500000)', 'Very High (>500000)'])
df['AMT_ANNUITY_BINS'] = pd.qcut(df['AMT_ANNUITY'], 5, labels=['Very Low(<20000)', 'Low (20000-40000)', 'Mid (40000-50000)', 'High (50000-60000)', 'Very High (>60000)'])
```

```python
# defining a function to plot seaborn distplots for Bivariate analysis with respect to "NAME_YIELD_GROUP"

def pre_bivar(col):
    plt.figure(figsize=(8,6))
    
    sns.distplot(approved[col], hist=False, label='Approved')
    sns.distplot(canceled[col], hist=False, label='Cancelled')
    sns.distplot(refused[col], hist=False, label='Refused')
    sns.distplot(unused_offer[col], hist=False, label='Unused')
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()
```

```python
# NAME_CONTRACT_STATUS Vs. AMT_ANNUITY
pre_bivar('AMT_ANNUITY')
```

```python
# defining a function to plot countplots for univariate categorical analysis

def prev_bivar_cat(col):
    plt.figure(figsize=(12,12))
    
    plt.subplot(2,2,1)
    plt.title('Approved')
    plt.xticks(rotation=45)
    sns.countplot(approved[col])
    
    plt.subplot(2,2,2)
    plt.title('Canceled')
    plt.xticks(rotation=45)
    sns.countplot(canceled[col])
    
    plt.subplot(2,2,3)
    plt.title('Refused')
    plt.xticks(rotation=45)
    sns.countplot(refused[col])
    
    plt.subplot(2,2,4)
    plt.title('Unused Offer')
    plt.xticks(rotation=45)
    sns.countplot(unused_offer[col])
    
    plt.show()
```
